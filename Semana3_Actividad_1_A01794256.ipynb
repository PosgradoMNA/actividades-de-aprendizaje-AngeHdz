#Parte 1: Fundamentos de bases de datos

1.   Fundamentos de bases de datos y para ciencia de datos.  
    Una base de datos es una recopilacion organizada de informacion o datos estructurados, que nomarmelmete se almacena de forma electronica en un sistema informatico. Normalmente este provee un denominado Freamwork que provee la estrucura, almacenamiento, y la recuperacion de datos, 
    Existen dos tipos de base de datos:

*   Relacionales (PostgresSQL, MySql, DB2)
    En la basases de datos relaciones la informacion esta organizada en tablas con elementos que pueden estar relacionados a otras tablas, la forma de interactuar con estas es usan el lenguaje SQL
*   No relacionales (RabbitMQ, MongoDB, JanusGraphs)
    Mejor concidas como NoSQL, las bases de datos no relacionales almacenan la informacion por categorias en lugar de relaciones. La forma mas popular de almacenamiento es utilizando Kay-Value muy similar a como lo hacen los diccionarios de python.Con una llave unica que da acceso a un denominado documento el cual esta relacionada a la llave. Se puede agregar, eliminar y realizar updates, siguiendo la estructura del documento.

2.   Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos
   Un Data Warehouse, es un systema que esta diseñado para almacenar datos y el analizis de estos, utilizando busines intelligence. Este sistema centraliza multiples datos, como datos en tiempo real, datos historicos y es conciderado com ola unica fuente de la verdad.
